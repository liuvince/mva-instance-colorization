```
import os
import moviepy.video.io.ImageSequenceClip

image_folder = 'result/colorization/colorization_IRT1_initial1/car-roundabout_input/0050'
video_dir = 'predict_color_dir'
fps = 20

folder="0050"

if not os.path.exists(video_dir):
  os.mkdir(video_dir)

image_files = sorted([image_folder+'/'+img for img in os.listdir(image_folder) if img.endswith(".jpg") and img.startswith("out_main_")])
clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)

clip.write_videofile(os.path.join(video_dir, folder + '.mp4'))
```
