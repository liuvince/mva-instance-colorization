```
import os
import moviepy.video.io.ImageSequenceClip

out = "0025"
video_dir = 'video_prior_epoch_{}'.format(out)
fps = 20

if not os.path.exists(video_dir):
  os.mkdir(video_dir)

for folder in os.listdir('dataset_grayscale'):
  image_folder = 'result/colorization/colorization_IRT1_initial1/{}/{}'.format(folder, out)
  image_files = sorted([image_folder+'/'+img for img in os.listdir(image_folder) if img.endswith(".jpg") and img.startswith("out_main_")])
  clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(image_files, fps=fps)

  clip.write_videofile(os.path.join(video_dir, folder + '.mp4'))
```

```
import os
import moviepy.video.io.ImageSequenceClip
import cv2
import numpy as np 

video_dir = 'video_prior_epoch_concat'
fps = 20

if not os.path.exists(video_dir):
  os.mkdir(video_dir)

tmp = "tmp"
if not os.path.exists(tmp):
  os.mkdir(tmp)

for folder in os.listdir('dataset_grayscale'):
  if not os.path.exists(os.path.join(tmp, folder)):
    os.mkdir(os.path.join(tmp, folder))
  
  image_files_per_epoch = []
  image_folder = os.path.join('dataset_grayscale', folder)
  image_files = sorted([image_folder+'/'+img for img in os.listdir(image_folder) if img.endswith(".jpg")])
  image_files_per_epoch.append(image_files)
  for out in ["0005", "0010", "0025"]:
    image_folder = 'result/colorization/colorization_IRT1_initial1/{}/{}'.format(folder, out)
    image_files = sorted([image_folder+'/'+img for img in os.listdir(image_folder) if img.endswith(".jpg") and img.startswith("out_main_")])
    image_files_per_epoch.append(image_files)
  
  frames = []
  for image_files in zip(*image_files_per_epoch):
    imgs = []
    for image_file in image_files:
      imgs.append(cv2.imread(image_file))
    vis = np.concatenate(imgs, axis=1)

    frames.append(os.path.join(os.path.join(tmp, folder), os.path.split(image_files[0])[-1])
    cv2.imwrite(frames[-1], vis)
    
  clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(frames, fps=fps)

  clip.write_videofile(os.path.join(video_dir, folder + '.mp4'))
```
